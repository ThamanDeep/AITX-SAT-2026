# autoresearch analysis

Plot `results.tsv` the way karpathy/autoresearch plots experiment progress.


```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('results.tsv', sep='\t')
print(df.tail())
kept = df[df.status.isin(['keep','baseline'])].copy()
disc = df[df.status == 'discard']

fig, ax = plt.subplots(figsize=(10, 5))
ax.scatter(disc.index, disc.accuracy, c='#c8c8c8', s=16, label='Discarded')
ax.scatter(kept.index, kept.accuracy, facecolors='none', edgecolors='#2e8b57', s=55, label='Kept')
best = None; xs=[]; ys=[]
for i, row in df.iterrows():
    if row.status in ('keep','baseline'):
        best = row.accuracy
    if best is not None:
        xs.append(i); ys.append(best)
ax.step(xs, ys, where='post', color='#2e8b57', lw=2, label='Running best')
ax.set_xlabel('Experiment #'); ax.set_ylabel('Accuracy')
ax.set_title('Autoresearch progress'); ax.legend(); ax.grid(True, alpha=.3)
plt.show()
```
